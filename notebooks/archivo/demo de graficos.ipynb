# Graficos Seaborn

## 1. GrÃ¡fico de dispersiÃ³n (Scatter Plot)
Para ver la relaciÃ³n entre dos variables numÃ©ricas.
Ejemplo: Distance (km) vs. Delivery Duration (min)
Puedes agregar la lÃ­nea de regresiÃ³n para visualizar el ajuste:

```python
import seaborn as sns
sns.regplot(x='Distance (km)', y='Delivery Duration (min)', data=df)
```

## 2. GrÃ¡fico de barras (Bar Plot)
Para analizar promedios o totales agrupados por categorÃ­as.
Ejemplo: Promedio de Tip ($) por tipo de taco (Taco Type)
TambiÃ©n Ãºtil para ver cÃ³mo cambian los valores segÃºn si fue un pedido de fin de semana.

```python
sns.barplot(x='Taco Type', y='Tip ($)', data=df)
```

## 3. Boxplot
Muestra la distribuciÃ³n y los valores atÃ­picos (outliers).
Ejemplo: DistribuciÃ³n de Price ($) por tamaÃ±o de taco.

```python
sns.boxplot(x='Taco Size', y='Price ($)', data=df)
```

## 4. Histogramas
Para entender la distribuciÃ³n de una variable numÃ©rica.
Ejemplo: Â¿CÃ³mo se distribuye la duraciÃ³n de entrega?

```python
df['Delivery Duration (min)'].hist(bins=20)
```

## 5. Heatmap de correlaciÃ³n
Ideal para ver cÃ³mo se relacionan todas las variables numÃ©ricas entre sÃ­.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only use numeric columns for correlation
corr = df.select_dtypes(include=['number', 'bool']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
