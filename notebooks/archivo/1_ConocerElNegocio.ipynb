# **Este Codigo**
# **Es el modelo Pipeline que mostrata el conjunto de procesos o etapas de este Negocio**

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- Datos del pipeline ---
data_pipeline = {
    'Captura del Dato': [
        'Ubicaci√≥n del cliente',
        'Tipo y tama√±o de taco',
        'Cantidad de toppings',
        'Tiempo de entrega',
        'Distancia recorrida',
        'Propina entregada',
        'Patrones de d√≠a'
    ],
    'An√°lisis Aplicado': [
        'Segmentaci√≥n geogr√°fica',
        'An√°lisis de preferencias',
        'Control de costos',
        'Evaluaci√≥n de eficiencia',
        'Optimizaci√≥n de rutas',
        'Medici√≥n de satisfacci√≥n',
        'An√°lisis comparativo'
    ],
    'Decisi√≥n Estrat√©gica': [
        'Planificaci√≥n de marketing',
        'Ingenier√≠a de men√∫',
        'Ajuste de precios',
        'Mejora de operaciones',
        'Reducci√≥n de gastos',
        'Programas de lealtad',
        'Gesti√≥n de demanda'
    ]
}

# --- Configuraci√≥n del gr√°fico ---
fig, ax = plt.subplots(figsize=(18, 8))
fig.patch.set_facecolor('#F1F3F5')  # Fondo general suave

# Colores para etapas
etapa_colores = ['#0D47A1', '#1976D2', '#42A5F5']  # Gama profesional azul

# Posiciones horizontales de las etapas
x_pos = [0.15, 0.5, 0.85]
etapas = list(data_pipeline.keys())
y_pos = 0.5

# Estilo de las cajas
box_style = dict(boxstyle='round,pad=0.5', fc='white', ec='#CED4DA', lw=1.8)
arrow_style = dict(arrowstyle='-|>', color='#343A40', lw=2, mutation_scale=18)

# Dibujar cajas y t√≠tulos
for i, etapa in enumerate(etapas):
    x = x_pos[i]
    items = '\n'.join(f'‚Ä¢ {line}' for line in data_pipeline[etapa])

    # Texto de la etapa
    ax.text(x, y_pos, items,
            ha='center', va='center',
            fontsize=11, color='#495057',
            bbox=box_style, linespacing=1.5)

    # T√≠tulo superior de cada etapa
    ax.text(x, 0.9, f'{i+1}. {etapa}',
            ha='center', va='center',
            fontsize=15, fontweight='bold',
            color=etapa_colores[i])

# Flechas entre etapas
for i in range(len(x_pos) - 1):
    arrow = patches.FancyArrowPatch(
        (x_pos[i] + 0.13, y_pos),
        (x_pos[i+1] - 0.13, y_pos),
        connectionstyle="arc3,rad=0.2",
        **arrow_style
    )
    ax.add_patch(arrow)

# Formato final
ax.axis('off')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Eliminar la lupa y modificar el t√≠tulo
plt.suptitle('Pipeline de Inteligencia de Negocio ‚Äì Taco Sales',
             fontsize=22, fontweight='bold', color='#212529', y=0.98)

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()

```

## Se mostrara la linea de tiempo historica del negocio

La siguiente l√≠nea de tiempo muestra los hitos m√°s relevantes del desarrollo del modelo Taco Sales, desde su conceptualizaci√≥n hasta su proyecci√≥n de expansi√≥n a Chile.


```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Datos de los eventos
eventos = {
    "Fecha": ["2020-01-01", "2021-06-15", "2022-03-10", "2023-08-05", "2024-01-01", "2025-06-01"],
    "Hito": [
        "Inicio del negocio", "Primera sucursal en Austin", "Sistema de pedidos",
        "Expansi√≥n a otras ciudades", "Optimizaci√≥n log√≠stica", "Evaluaci√≥n del mercado chileno"
    ],
    "Fase": ["Inicio", "Validaci√≥n", "Tecnolog√≠a", "Expansi√≥n", "Optimizaci√≥n", "Evaluaci√≥n"]
}

# Crear DataFrame
df = pd.DataFrame(eventos)
df["Fecha"] = pd.to_datetime(df["Fecha"])

# Colores para cada fase
colores = {"Inicio": "#FFB74D", "Validaci√≥n": "#4CAF50", "Tecnolog√≠a": "#2196F3",
           "Expansi√≥n": "#FF5722", "Optimizaci√≥n": "#9C27B0", "Evaluaci√≥n": "#00BCD4"}

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(10, 5))
for i, row in df.iterrows():
    ax.plot(row["Fecha"], 1, "o", color=colores[row["Fase"]], markersize=10)
    ax.text(row["Fecha"], 1.1, row["Hito"], ha='center', va='bottom', fontsize=9, rotation=45)

ax.set_title("L√≠nea de Tiempo - Avance del Negocio", fontsize=14)
ax.set_xticks(pd.to_datetime(df["Fecha"]))
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.get_yaxis().set_visible(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
