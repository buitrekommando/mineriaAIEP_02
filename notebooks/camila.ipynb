# Convertir la columna Order Time a datetime

```python
df['Order Time'] = pd.to_datetime(df['Delivery Time'], dayfirst=True)
```

# Separar de order time, la fecha y hora en columnas diferentes

```python
df['Order Date'] = df['Order Time'].dt.date
df['Order Hour'] = df['Order Time'].dt.hour
df['Order Minute'] = df['Order Time'].dt.minute
df['Order Hour Decimal'] = df['Order Hour'] + df['Order Minute'] / 60
```

```python
df
```

# Se quita order time de la lista 

```python
df.drop(columns=['Order Time'], inplace=True)
```

# Mostramos df

```python
df
```

 # Separar de Delivery Time, la fecha y hora en columnas diferentes

```python
df['Delivery Time'] = pd.to_datetime(df['Delivery Time'], dayfirst=True)
df['Delivery Date'] = df['Delivery Time'].dt.date
df['Delivery Hour'] = df['Delivery Time'].dt.time
```

# Dejamos el campo hora como número decimal

```python
df['Delivery Hour Decimal'] = df['Delivery Time'].dt.hour + df['Delivery Time'].dt.minute / 60
```

# Eliminamos la colunma original

```python
df.drop(columns=['Delivery Time'], inplace=True)
```

# Mostramos df

```python
df
```
