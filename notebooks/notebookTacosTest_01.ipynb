# Graficos Seaborn

## 1. Gráfico de dispersión (Scatter Plot)
Para ver la relación entre dos variables numéricas.
Ejemplo: Distance (km) vs. Delivery Duration (min)
Puedes agregar la línea de regresión para visualizar el ajuste:

```python
import seaborn as sns
sns.regplot(x='Distance (km)', y='Delivery Duration (min)', data=df)
```

## 2. Gráfico de barras (Bar Plot)
Para analizar promedios o totales agrupados por categorías.
Ejemplo: Promedio de Tip ($) por tipo de taco (Taco Type)
También útil para ver cómo cambian los valores según si fue un pedido de fin de semana.

```python
sns.barplot(x='Taco Type', y='Tip ($)', data=df)
```

## 3. Boxplot
Muestra la distribución y los valores atípicos (outliers).
Ejemplo: Distribución de Price ($) por tamaño de taco.

```python
sns.boxplot(x='Taco Size', y='Price ($)', data=df)
```

## 4. Histogramas
Para entender la distribución de una variable numérica.
Ejemplo: ¿Cómo se distribuye la duración de entrega?

```python
df['Delivery Duration (min)'].hist(bins=20)
```

## 5. Heatmap de correlación
Ideal para ver cómo se relacionan todas las variables numéricas entre sí.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only use numeric columns for correlation
corr = df.select_dtypes(include=['number', 'bool']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
